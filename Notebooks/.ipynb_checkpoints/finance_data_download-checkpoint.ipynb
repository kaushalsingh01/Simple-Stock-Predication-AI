```python
import yfinance as yf
import pandas as pd
```

```python
# List of 10 major companies (customize if needed)
tickers = [
    "AAPL",  # Apple
    "MSFT",  # Microsoft
    "GOOGL", # Google
    "AMZN",  # Amazon
    "META",  # Meta (Facebook)
    "TSLA",  # Tesla
    "JPM",   # JPMorgan Chase
    "WMT",   # Walmart
    "NVDA",  # NVIDIA
    "V"      # Visa
]
```

```python
# Fetch 3 years of daily data (adjust dates if needed)
for ticker in tickers:
    data = yf.download(ticker, start="2021-01-01", end="2024-01-01")
    data.to_csv(f"{ticker}_stock.csv")  # Save as CSV
    print(f"✅ Saved {ticker} data to {ticker}_stock.csv")

print("Done! All 10 stocks downloaded.")
```
